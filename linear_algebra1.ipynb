# Linear algebra in python

```python
import numpy as np
```

2x+3y = 8 

5x+4y = 20

```python
# coefficient matrix A
A=np.array([[2,3],[5,4]])
A
```

```python
# constant matrxi B
B=np.array([8,20])
```

```python
#solve for x
x=np.linalg.solve(A,B)
print(x)
y=np.linalg.solve(A,B)
print(y)
```

# Solving Complex Equations

\[ 
\begin{align*}
(1 + 2i)x + (2 - i)y &= 4 + i \\
(3 - i)x + 4y &= 3 + 4i
\end{align*} 
\]

```python
# Coefficients matrix A with complex numbers
A=np.array([[1+2j , 2-1j],[3-1j,4]])
print(A)
B=np.array([4+1j,3+4j])
print(B)
x=np.linalg.solve(A,B)
x
```

# SVD in Python

# Vectors in Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#crate a vector
v=np.array([2,3])
v
```

```python
# plot the vector
plt.figure()
plt.quiver(0,0,v[0],v[1],scale_units="xy",angles="xy",scale=1,color="green")
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.title("2D vector")
plt.xlabel("X")
plt.ylabel("Y")
plt.gca().set_aspect("equal",adjustable="box")
plt.show()
```

---
# Linear Transformation

\[ 
\begin{bmatrix}
\cos(45^\circ) & -\sin(45^\circ) \\
\sin(45^\circ) & \cos(45^\circ)
\end{bmatrix}
\]

```python
# Define the rotation matrix for 45 degree rotation
theta=np.radians(50)
rotation_matrix=np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
print(rotation_matrix)
print("---------------------")
print(v)
# applying the transformation
v_transformed=np.dot(rotation_matrix,v)
v_transformed
```

```python
# Plotting the original and transformed vectors
plt.figure()
plt.quiver(0,0,v[0],v[1],scale_units="xy",angles="xy",scale=1,color="green",label="Original Vector")
plt.quiver(0,0,v_transformed[0],v_transformed[1],scale_units="xy",angles="xy",scale=1,color="b",label="Transformed vector")
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid()
plt.title("2D vector transformation")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.gca().set_aspect("equal",adjustable="box")
plt.show()
```

```python
# Adjusting the plot to show the grid lines behind the vectors
plt.figure()
plt.quiver(0,0,v[0],v[1],scale_units="xy",angles="xy",scale=1,color="green",label="Original Vector")
plt.quiver(0,0,v_transformed[0],v_transformed[1],scale_units="xy",angles="xy",scale=1,color="b",label="Transformed vector")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='k',linewidth=0.5)
plt.axvline(x=0, color='k',linewidth=0.5)
plt.title('2D Vector Transformation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# Shear Transformation

```python
# Define the shear matrix for shearing along the x-axis with k = 1.5
shear_factor=2.5
shear_matrix=np.array([[1,shear_factor],[0,1]])
# applying the shear transformation
v_sheared=np.dot(shear_matrix,v)
# Plotting the original, rotated, and sheared vectors
plt.figure()
plt.quiver(0,0,v[0],v[1],scale_units="xy",angles="xy",scale=1,color="green",label="Original Vector")
plt.quiver(0,0,v_transformed[0],v_transformed[1],scale_units="xy",angles="xy",scale=1,color="b",label="Transformed vector")
plt.quiver(0,0,v_sheared[0],v_sheared[1],scale_units="xy",angles="xy",scale=1,color="r",label="Sheared vector")
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='k',linewidth=0.5)
plt.axvline(x=0, color='k',linewidth=0.5)
plt.title('2D Vector Transformation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
